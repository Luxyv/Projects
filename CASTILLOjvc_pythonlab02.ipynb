```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from scipy.stats import norm
```

```python
#A. The average are burned by forest fires in New Sourth Whales is 4,300 acres per year, with a standard deviation of 750 acres. The distribution of the number of acres burned is normal. What is the probability that between 2,500 and 4,200 acres will be burned in any given year? (30 points)
#given: average = 4300, std = 750

#create the distribution
data = np.arange((4300 - 3*750), (4300 + 3*750), 0.1) #create an array of evenly spaced numbers
pdf = norm.pdf(data, loc = 4300, scale = 750)

#visualize the normal curve
sb.set_style('darkgrid') #bg of the plot/chart
sb.lineplot(x = data, y = pdf, color = 'black')
plt.xlabel('acres burned')
plt.ylabel('probability density')
```

```python
#What is the probability that between 2,500 and 4,200 acres will be burned in any given year?
#P(2500 < x < 4200)
probA = norm(loc = 4300, scale = 750).cdf(2500)
probB = norm(loc = 4300, scale = 750).cdf(4200)
print (f"{probB - probA: .2%}")
```

```python
#B. The average salary for first-year teachers is 27, 989php. Assume the distribution is approximately normal with standard deviation 3,250php. What is the probability that a randomly selected first-year teacher makes between 20,000php and 30,000php each year? (30 points) (graph)
#Given: mean = 27,989  std = 3,250

#create the distribution
data = np.arange((27989 - 3*3250), (27989 + 3*3250), 0.1)
pdf = norm.pdf(data, loc = 27989, scale = 3250)

#visualize the normal curve
sb.set_style('darkgrid')
sb.lineplot(x = data, y = pdf, color = 'black')
plt.xlabel('salary')
plt.ylabel('probability density')
```

```python
#What is the probability that a randomly selected first-year teacher makes between 20,000php and 30,000php each year?
#P(20000 < x < 30000)
probA = norm(loc = 27989, scale = 3250).cdf(20000)
probB = norm(loc = 27989, scale = 3250).cdf(30000)
print (f"{probB - probA: .2%}")
```
